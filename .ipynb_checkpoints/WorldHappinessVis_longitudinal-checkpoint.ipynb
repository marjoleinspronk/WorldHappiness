## Visualization of the combined [World Happines Ranking 2017](https://www.kaggle.com/unsdsn/world-happiness), [The World Factbook Data](https://www.kaggle.com/fernandol/countries-of-the-world) and [The Global Religion 2010](https://www.kaggle.com/arthurtok/global-religion-1945-2010-plotly-pandas-visuals/data) datasets

#### NB: Note that the happiness score is based on the factors "GDP", "Social Support (Family)", "Life Expectation", "Freedom", "Generosity", and "Trust & Goverment Corruption" in that dataset, which describe the extent to which these factors contribute in evaluating the happiness in each country, so a prediction using these factors does not make sense.

#### Other measures we have that we can use from the other two datasets (whether it makes sense is a different question):
 * population
 * area
 * population density
 * coastline
 * net migration
 * literacy
 * phones
 * arable, crops
 * climate
 * birthrate
 * agriculture, industry, service
 * religion (here: percentage of population without religion)


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dfall_2017 = pd.read_csv('2017.csv')
dfall_2017.head()
```

```python
dfall_2017.columns = ['Country', 'HappinessRank', 'HappinessScore','WhiskerHigh','WhiskerLow','GDP',
              'Family','LifeExpect','Freedom','Generosity','TrustGovCorrup','DystopiaRes']
whiskers = ['WhiskerHigh','WhiskerLow' ]
df_17 = dfall_2017.drop(whiskers, axis=1)
df_17.head()
```

## Exploring the dataset  - some visualizations

#### The happiness scores were calculated for each country based on measured factors (columns in the dataset). For example, for Denmark:

```python
denmark = df_17.iloc[1]
happyscore = (denmark['GDP'] + denmark['Family'] + denmark['LifeExpect'] + denmark['Freedom'] +
              denmark['Generosity'] + denmark['TrustGovCorrup'] + denmark['DystopiaRes'])
print(happyscore)
```

#### World map with Happiness Ranking for the year 2017

```python
import plotly.plotly as py 
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)

data4plot = dict(type = 'choropleth', 
           locations = df_17['Country'],
           locationmode = 'country names',
           z = df_17['HappinessRank'], 
           text = df_17['Country'],
          colorscale = 'Hot', reversescale = True)
mapspec = dict(title = 'World Happiness Ranking 2017', 
             geo = dict(showframe = False, projection = {'type': 'Natural Earth'}))
HappyMap = go.Figure(data = [data4plot], layout=mapspec)
iplot(HappyMap)
```

#### Scatter plot showing the relation between GDP and Happiness

```python
# Here you can see how GDP and happiness ranking are related
# Note that GDP is the factor contributing most to the Happiness Score
gdp = df_17['GDP']
happyrank = df_17['HappinessRank'] 

plt.scatter(gdp,happyrank)
plt.gca().invert_yaxis()
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Ranking')
plt.title('Higher GDP --> Higher Ranking')
plt.text(1.616463, 1, 'Norway No.1')
plt.text(1.41692, 50, 'Japan No.50')
plt.text(1.0547, 100, 'South Africa No.100')

plt.show()
```

## Does money make happy?
#### There are some shifts in ranking when GDP is left out of the Happiness Score, but nothing major happens:

```python
# What happens to the ranking when we leave out the money factor?
# GDP is the biggest contributer to the Happiness Score

# Create a new column with the sum of all factors except GDP
df_17['HappinessScore_WithoutMoney']=0
for index, row in df_17.iterrows():
    newscore = row['Family'] + row['LifeExpect'] + row['Freedom'] + row['Generosity'] + row['TrustGovCorrup'] + row['DystopiaRes']
    df_17.loc[index, "HappinessScore_WithoutMoney"] = newscore

df_17['HappyRank_WithoutMoney'] = df_17['HappinessScore_WithoutMoney'].rank(ascending=False)
df_17.head()
```

```python
# Plot ranking without the GDP data
data4plot2 = dict(type = 'choropleth', 
           locations = df_17['Country'],
           locationmode = 'country names',
           z = df_17['HappyRank_WithoutMoney'], 
           text = df_17['Country'],
          colorscale = 'Hot', reversescale = True)
mapspec = dict(title = 'Happiness Ranking 2017 - without GDP data', 
             geo = dict(showframe = False, projection = {'type': 'Natural Earth'}))
HappyMap = go.Figure(data = [data4plot2], layout=mapspec)
iplot(HappyMap)
```

#### Plot GDP


```python
# Plot ranking without the GDP data

df_17['relativeGDP'] = df_17['GDP']/df_17['HappinessScore']

data4plot2 = dict(type = 'choropleth', 
           locations = df_17['Country'],
           locationmode = 'country names',
           z = df_17['relativeGDP'], 
           text = df_17['Country'],
          colorscale = 'Greens', reversescale = True)
mapspec = dict(title = 'GDP contributing factor data, corrected for tot happiness score', 
             geo = dict(showframe = False, projection = {'type': 'Natural Earth'}))
HappyMap = go.Figure(data = [data4plot2], layout=mapspec)
iplot(HappyMap)
```

## Exploring the merged data: Is there a correlation between Population Density and Happiness Score?

```python
popdensity = df_merged_rel['PopDensity']
happyscore = df_merged_rel['HappinessScore'] 
plt.scatter(popdensity,happyscore)
plt.xlabel('Population Density')
plt.ylabel('Happiness Score')
plt.title('Hmmm... seems like we have an outlier')
plt.show()
```

```python
# Remove very densily populated Singapore from the data
idx_outlier = df_merged_rel['PopDensity'] < 4000
df_merged_nooutlier = df_merged_rel[idx_outlier]
```

```python
# Is there a correlation between Population Density and Happiness Score?
popdensity = df_merged_nooutlier['PopDensity']
happyscore = df_merged_nooutlier['HappinessScore'] 
plt.scatter(popdensity,happyscore)
#plt.gca().invert_yaxis()
plt.xlabel('Population Density')
plt.ylabel('Happiness Score')

plt.title('No significant relation between pop. density and happiness')
plt.show()
```

#### Let's calculate a correlation score and p-value

```python
import scipy.stats  as stats

# Looking at the plot, this correlation isn't expected to be significant...
stats.pearsonr(df_merged_nooutlier['PopDensity'], df_merged_nooutlier['HappinessScore'])   
```

### What about Infant mortality and happiness? Here we find a significant correlation
#### However, what do the factors 'family' and 'life expectancy' represent on which the happiness score is based? Infant mortality is likely to be part of these factors as well, so this significant negative correlation is not surprising...

```python
from matplotlib import colors as mcolors

colors = list(df_new['Color'].unique())
regions = list(df_new['Region'].unique())

for i in range(0 , len(colors)):
    data = df_new.loc[df_new['Color'] == colors[i]]
    plt.scatter('InfantMort', 'HappinessScore', data=data, color='Color', label=regions[i], alpha = 0.8)

plt.xlabel('Infant Mortality')
plt.ylabel('Happiness Score')
plt.title('Relation infant mortality and happiness')
#plt.figure(figsize=(20,10))


plt.legend(bbox_to_anchor=(1, 1))
plt.show()
```

```python
stats.pearsonr(df_merged_nooutlier['InfantMort'], df_merged_nooutlier['HappinessScore']) 
```

```python
colors = list(df_new['Color'].unique())
regions = list(df_new['Region'].unique())

plt.figure(figsize=(10,8))
for i in range(0 , len(colors)):
    data = df_new.loc[df_new['Color'] == colors[i]]
    datamean_rel = data['noreligion_percent'].mean()
    datamean_happy = data['HappinessRank'].mean()  
    #plt.scatter('noreligion_percent', 'HappinessScore', data=data, color='Color', label=regions[i], alpha = 0.8)
    #plt.bar(i, datamean_happy, width=0.8, align='center', label=regions[i])
    plt.bar(datamean_happy, datamean_rel, width=5, align='center', label=regions[i], alpha=0.8)


plt.xlabel('Mean Happiness Ranking')
plt.ylabel('Mean % of pop. without religion')
plt.title('')
#plt.figure(figsize=(20,10))

plt.legend(bbox_to_anchor=(1, 1))
plt.show()
```

```python
# Normalize population size for plotting
maxval = df_new['Population'].max()   #1313973713
minval = df_new['Population'].min()   # 287730

df_new['popnorm'] = (df_new['Population'] - minval) / (maxval - minval) * 2000
```

```python
colors = list(df_new['Color'].unique())
regions = list(df_new['Region'].unique())


plt.figure(figsize=(10,8))
for i in range(0 , len(colors)):
    data = df_new.loc[df_new['Color'] == colors[i]] 
    plt.scatter('noreligion_percent', 'HappinessScore', data=data, color='Color', s='popnorm', label=regions[i], alpha = 0.8)


plt.xlabel('Mean Happiness Ranking')
plt.ylabel('Mean % of pop. without religion')
plt.title('')
#plt.figure(figsize=(20,10))

plt.legend(bbox_to_anchor=(1, 1))
plt.show()
```
